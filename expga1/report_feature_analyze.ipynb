## Exploración de la selección de características

```python
# Replace 'your_file.json' with the path to your JSON file
with open('experiments.json', 'r') as file:
    features = json.load(file)
```

```python
# for i in features:
#     print(i)
#     print(features[i]['selected_features_bin'])
#     print('-'*10)
```

```python
all_features_selected_original = []
all_features_selected_augmented = []
for i in features:    
    if  '0001' in i:
            all_features_selected_original += features[i]['selected_features_bin']
    elif '0002' in i:
          all_features_selected_augmented  += features[i]['selected_features_bin']
                  
```

Conjunto de features seleccionados en los 30 experimentos al menos una vez

```python
len(set(all_features_selected_augmented))
```

```python
len(set(all_features_selected_original))
```

Los features están muy correlacionados. Con el 10% de los datos se resuelve el problema, y no importa cuál 10%?

```python
features_original = pd.DataFrame({'features_original' : all_features_selected_original}) \
    .groupby('features_original') \
    .size() \
    .reset_index(name='freq') \
    .sort_values(by='freq', ascending=False) \
    .rename(columns={'features_original': 'features'})
```

```python
features_augmented = pd.DataFrame({'features_original' : all_features_selected_augmented}) \
    .groupby('features_original') \
    .size() \
    .reset_index(name='freq') \
    .sort_values(by='freq', ascending=False) \
    .rename(columns={'features_original': 'features'})
```

## Distribución de frecuencia de los features seleccionados

```python
plt.hist(features_original['freq'])
plt.xlabel('freq')
plt.ylabel('observations') 
plt.title('Frecuency of features selected across trials in Original Dataset')
```

### Evaluación de los features seleccionados considerando los dos datasets 

```python
merged_features = pd.merge(features_original_top100, features_augmented_top100, on='features', how='outer', suffixes=('_df1', '_df2'))
```

```python
df_freq = pd.merge(features_original_top100, features_augmented_top100, on='features', how='outer', suffixes=('_df1', '_df2')) \
    .fillna(0) \
    .assign(freq=lambda x: x['freq_df1'] + x['freq_df2']) \
    .groupby('features', as_index=False)['freq'] \
    .sum() \
    .sort_values(by='freq', ascending=False)

print(df_freq)
```
